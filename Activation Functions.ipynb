```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import keras
import tensorflow as tf
```

```python
a=tf.constant([-4,-3,-2,-1,0,1,2,3,4],dtype=tf.float32)
b=keras.activations.sigmoid(a)
b.numpy()
```

```python
a=tf.constant([-4,-3,-2,-1,0,1,2,3,4],dtype=tf.float32)
b=keras.activations.relu(a)
b.numpy()
```

```python
a=tf.constant([-4,-3,-2,-1,0,1,2,3,4],dtype=tf.float32)
b=keras.activations.selu(a)
b.numpy()
```

```python
a=tf.constant([-4,-3,-2,-1,0,1,2,3,4],dtype=tf.float32)
b=keras.activations.elu(a)
b.numpy()
```

```python
a=tf.constant([-4,-3,-2,-1,0,1,2,3,4],dtype=tf.float32)
b=keras.activations.tanh(a)
b.numpy()
```

```python
a=tf.constant([[-2,-1,0,1,2]],dtype=tf.float32)
b=keras.activations.softmax(a,axis=1)
b.numpy()
```

```python
a=tf.constant([[-2,-1,0,1,2]],dtype=tf.float32)
b=keras.activations.softplus(a)
b.numpy()
```
